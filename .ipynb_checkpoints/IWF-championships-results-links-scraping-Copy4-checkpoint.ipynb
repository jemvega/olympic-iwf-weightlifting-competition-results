```python
import re
import requests
from bs4 import BeautifulSoup
import parsing_functions as pf
import time
import concurrent.futures
```

```python
url_domain = "https://en.wikipedia.org"
website_url = "https://en.wikipedia.org/wiki/World_Weightlifting_Championships"
header_name = "Combined"

competition_urls = [url_domain + elem for elem in pf.WikiParser.wikitable_links(website_url, header_name, pf.years_list)]

def process_urls(url):
    website_url = requests.get(url)
    page = BeautifulSoup(website_url.text, "lxml")
    results_urls = []
    for link in page.find_all("a", attrs={"href": re.compile(f"^{url[24:]}_")}):
        results_urls.append(link.get('href'))
    return results_urls

with concurrent.futures.ThreadPoolExecutor() as executor:
    weightclass_urls = list(executor.map(process_urls, competition_urls))
    
comp_urls = []
for index in range(len(weightclass_urls)):
    for elem in weightclass_urls[index]:
        comp_urls.append(url_domain + elem) 
comp_urls
```

```python
with concurrent.futures.ThreadPoolExecutor() as executor:
    for elem in comp_urls:
        pf.datatable_cleanup.results_table(elem)
```
