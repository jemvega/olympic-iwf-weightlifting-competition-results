```python
def check_float(string):
    try:
        float(string)
        return True
    except ValueError:
        return False
```

```python
def string_to_string(converted_list):
    for elem in converted_list:
        if check_float(elem) == False:
            index = converted_list.index(elem)
            converted_list.pop(index)
            converted_list.insert(index, "NaN")
            return converted_list
```

```python
"""Question that remains: Will it be better to change the ___ entries to 
'NaN', 0, -999, '', or 'null'
Converting to 0 implies that no weight was lifted. This might be my best option. 
Converting to 'NaN' does not allow me as much versatility as a floating number when using aggregate functions. 
I'm not sure whether -999 might ruin stats in aggregate functions where 0 does not. 
I cannot sort or use aggregate functions with strings like '' and 'null'. 
As a result, I'm going to settle on 0 for now. """
def string_to_float(converted_list):
    for elem in converted_list:
        if check_float(elem):
            rational = float(elem)
            index = converted_list.index(elem)
            converted_list.pop(index)
            converted_list.insert(index, rational)
        elif check_float(elem) == False:
#             new_elem = "NaN"
#             floating = float(new_elem)
            index = converted_list.index(elem)
            converted_list.pop(index)
            converted_list.insert(index, 0)
        else:
            print("error")
    return converted_list
```
